```python
import sqlite3 as db
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
conn = db.connect("/Users/timhulak/Desktop/Syracuse/MBC-638/Project/DB/AML.db")
```

```python
data = pd.read_sql("""

                    SELECT 
                            "Reason"
                            ,"Stock"
                            ,"Year"
                            ,"Make"
                            ,"Model"
                            ,"O-D Lane"
                            ,"O-D Lead Time"
                            ,"Origin Name"
                            ,"Origin Region"
                            ,"Destination Name"
                            ,"Destination Region"
                            ,"Current Leg Lane"
                            ,"Leg Lead Time"
                            ,"Current/Most Recent Location"
                            ,"Next Location"
                            ,"Carrier"
                            ,"Movement Status"
                            ,"Hub Status"
                            ,"Approve Date"
                            ,"Original Dispatch Date"
                            ,"Current Leg Dispatch Date"
                            ,"Current Leg Due Date"
                            ,"Final Due Date"
                            ,"A-State AS Origin State"
                            ,"Days Late to Store"
                            ,"Days Late in Lane"
                            ,"Days Past Leg Tender Date"
                            ,"consolidated Carrier Name"
                            ,"AML_Date"
                    FROM AllMoves 
                    WHERE Reason = 'CUSTOMER_XFER'
                    
                    """,
                    con=conn,
                   parse_dates=['Approve Date',
                                'Original Dispatch Date',
                                'Current Leg Dispatch Date',
                                'Current Leg Due Date',
                                'Final Due Date',
                                'AML_Date'])
```
